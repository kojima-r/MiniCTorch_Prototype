```python
import sys
sys.path.append("../")

import json
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import minictorch.generator as GN
import minictorch.converter as CV
```

ニューラルネットの定義

```python
class Net(nn.Module):
   def __init__(self):
      super(Net, self).__init__()
      self.t = torch.tensor(np.array([[1.0, 2], [3, 4]]))
           
   def forward(self, x):
      f1 = 10 * x * x * self.t
      f2 = 5 * x
      y = f1 + f2
      return y
```

プロジェクト名、作業フォルダ、JSONファイル名の設定

```python
project = "test1"
folder = "./test"
json_path = folder + '/' + project +'.json'
```

計算グラフをJSONファイルに出力する

```python
def generate_json( json_path, input_x ):
   model = Net()
   model.eval()
   with torch.no_grad():
      GN.generate_minictorch_file( model, input_x, json_path )
   return model
```

```python
x = torch.tensor(np.array([[1.0, 2], [3, 4]]), requires_grad=True)
print(x)

model = generate_json( json_path, x )
```

JSONファイルをC++コードに変換する

```python
CV.convert_cpp_code( project, folder, model, x, json_path )
```

pythonコードを実行する

```python
# check code
model.train()
y = model( x )
print(y)
y.sum().backward()
print(x.grad)
```

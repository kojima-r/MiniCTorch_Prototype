```python
import sys
sys.path.append("../")

import json
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import minictorch.generator as GN
import minictorch.converter as CV
```

ニューラルネットの定義

```python
class Net( nn.Module ):
   def __init__(self):
      super(Net, self).__init__()
      self.fc1 = nn.Linear(2, 32)
      self.fc2 = nn.Linear(32, 32)
      self.fc3 = nn.Linear(32, 3)

   def forward(self, x):
      x = F.relu(self.fc1(x))
      x = F.relu(self.fc2(x))
      x = self.fc3(x)
      return x
      #return F.log_softmax(x, dim = 1)
```

プロジェクト名、作業フォルダ、JSONファイル名の設定

```python
project = "test2"
folder = "./test"
json_path = folder + '/' + project +'.json'
```

計算グラフをJSONファイルに出力する

```python
def generate_json( json_path, input_x ): 
    model = Net()
    print(input_x)

    model.eval()
    with torch.no_grad():
        print("[JSON]", json_path )
        GN.generate_minictorch_file( model, input_to_model, json_path )
    return model
```

```python
torch.manual_seed( 1 )
input_to_model = torch.randn((1,2),requires_grad=True)

model = generate_json( json_path, input_to_model )
```

JSONファイルをC++コードに変換する

```python
CV.convert_cpp_code( project, folder, model, input_to_model, json_path )
```

pythonコードを実行する

```python
# check code
#torch.set_grad_enabled(True)  #　効かない(?)
yy = model( input_to_model )
print(yy)
yy.sum().backward()
print(input_to_model.grad)
```

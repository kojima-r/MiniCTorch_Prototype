```python
import sys
sys.path.append("../")

import json
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import minictorch.generator as GN
import minictorch.converter as CV
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.t = torch.tensor(np.array([[5.0,6,7], [8,9,10]]))
        #self.b = torch.tensor([1.0,2.,3.],dtype=torch.float64)

    def forward(self, x):
        tt=torch.t(self.t)
        z1=torch.matmul(x,tt)
        print(z1)
        #z2=torch.sigmoid(z1)
        z2=torch.relu(z1)
        #z2 = torch.tanh(z1)
        #fn = torch.nn.ELU()
        #fn = torch.nn.LeakyReLU()
        #fn = torch.nn.ReLU6()
        #z2 = fn(z1)
        #z2,_= torch.max( z1, 0 )
        #print(z1)
        #print(z2)
        return z2
```

```python
def generate_json( json_path, input_x ):

    model = Net()
    model.eval()
    with torch.no_grad():
        print("[JSON]", json_path )
        GN.generate_minictorch_file( model, input_x, json_path )
        
    return model
```

## ニューラルネットの定義

```python
class Net(torch.nn.Module):
   def __init__(self, a):
      super(Net, self).__init__()
      self.x = a

   def forward(self,y):
      z = torch.matmul(self.x, y)
      return z
```

##プロジェクト名、作業フォルダ、JSONファイル名の設定

```python
project = "test05"
folder = "./test"
json_path = folder + '/' + project +'.json'
```

```python
! mkdir -p test
```

## 計算グラフをJSONファイルに出力する


```python
def generate_json( json_path, input_x, input_y ):
   model = Net( input_x )
   model.eval()
   with torch.no_grad():
      minictorch.trace( model, input_y, json_path )
   return model
```

```python
a=np.array( [[[1., 2, 3, 4],[4, 5, 6, 7],[7, 8, 9,10]],
            [[1, 2, 3, 4],[4, 5, 6, 7],[7, 8, 9, 9]]] )
b=np.array([[1., 2, 3],[4, 5, 6],[4, 5, 6],[7, 8, 9]])
x = torch.tensor( a )
y = torch.tensor( b , requires_grad=True)
print("a",a)
print("b",b)
model = generate_json( json_path, x, y )
```

## JSONファイルをC++コードに変換する

```python
minictorch.convert_all( project, folder, model, json_path, y, {"inp_data":y}, code="test" )
```

## pythonコードを実行する

```python
# check code
model.train()
z = model( y )
print(z)
z.sum().backward()
print(x.grad)
print(y.grad)
```
